# Data Cleaning

```python
data.drop(['bedrooms','neighbourhood'] , axis=1 , inplace=True)
```

```python
data['price'] = data['price'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
cat_col=data.select_dtypes(include='O').columns
print(list(cat_col))
```

```python
num_col=data.select_dtypes(exclude='O').columns
print(list(num_col))
```

```python
dat= data[data.isna()].columns

for x in dat:
    if data[x].dtype == cat_col:
        mode_value = data[x].mode()[0]
        data[x].fillna(mode_value, inplace=True)
    elif data[x].dtype == 'int64' or data[x].dtype == 'float64':
        mean_value = data[x].median()
        data[x].fillna(mean_value, inplace=True)
data.isna().sum()
```

# Identify the outliers

```python
#max
np.max(data.price)
```

```python
#min
np.min(data.price)
```

```python
#Show Outliers
sns.boxplot(data=data.price)
plt.title('Show Outliers For Price',fontsize=20)
plt.ylabel('Value',fontsize=20)
plt.xlabel('Price',fontsize=20)
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
q75,q25 = np.percentile(data.price,[75,25])
intr_qr = q75-q25
max = q75+(1.5*intr_qr)
min = q25-(1.5*intr_qr)
q75 , q25 , intr_qr , max , min
```

```python
data.price[data.price>max]=max
data.price[data.price<min]=min
#Show outliers
sns.boxplot(data=data.price)
plt.title('Outliers For Price',fontsize=20)
plt.ylabel('Value',fontsize=20)
plt.xlabel('Price',fontsize=20)
plt.grid(True)
plt.show()
```
